Iniciaizadores Avanzados

```python
# Zero matrix
m2 = np.zeros([5,4, 3])
print(m2.shape)

m3 = 2.0 * np.ones([5,4, 3])
#print(m3)

m4 = 10.0 * (np.random.rand(5,4) - 0.5)
print(m4)
```

```python
# Get elements
print("Element 0,1 m4: ", m4[0,1])

# Replace elements
m4[0,1] = 100.0
print("Element 0,1 m4: ", m4[0,1])

# Replace column
m4[:,0] = 2.0
print(m4)

m4[0:3,1] = 3.0
print(m4)

m4[3,:] = 4.0
print(m4)
```

## Operaciones en numpy

```python
A = np.random.rand(4,4)
B = np.random.rand(4,4)
V = np.random.rand(4,1)

print("A: " + str(A))
print("B: " + str(B))
print("V: " + str(V))

C = A*B # Element-wise multiplication
D = np.dot(A,B) # Matrix multiplication
E = np.dot(A,V) + V + 2 # Broadcasting
F = A + V
print("A*B = " + str(C))
print("A.B = " + str(D))
print("A.V + V = " + str(E))
print("A + V = " + str(F))
```
