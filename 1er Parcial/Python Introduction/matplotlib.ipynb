# 3. Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
y2 = np.cos(x)

plt.figure(0, figsize=(5, 2))
plt.plot(x, y, 'r', label='sin(x)')
plt.plot(x, y2, 'b', label='cos(x)')
plt.legend(loc='best')
plt.grid()
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Make arrays graphic

```python
plt.figure(2, figsize=(5, 2))
plt.subplot(2,2,1)
plt.plot(x, y, 'r')
plt.subplot(2,2,2)
plt.plot(x, y2, 'b')
plt.subplot(2,2,3)
plt.plot(x, y, 'c')
plt.subplot(2,2,4)
plt.plot(x, y2, 'y')
plt.show()
```

### scatter graphics (2D)

```python
# Create array
x = np.zeros([400,2])
y = np.zeros(400)

# Create first cluster
x[0:200,0] = 0.4 + 0.15*np.random.rand(200) - 0.15*np.random.rand(200)
x[0:200,1] = 0.4 + 0.15*np.random.rand(200) - 0.15*np.random.rand(200)
y[0:200] = np.zeros(200)

# Create second cluster
x[200:,0] = 0.6 + 0.15*np.random.rand(200) - 0.15*np.random.rand(200)
x[200:,1] = 0.6 + 0.15*np.random.rand(200) - 0.15*np.random.rand(200)
y[200:] = np.ones(200)

# Plot data
plt.figure(3, figsize=(5, 5))
plt.scatter(x[:,0], x[:,1], c=y, marker='o', s=40, cmap=plt.cm.Spectral)
```

### 3D Graphs (PLotly)

```python
# Import plotly
import plotly.graph_objects as go

# Create figure
fig = go.Figure(
    data=[
        go.Scatter3d(
        x = x[0:200,0],
        y = x[0:200,1],
        z = y[0:200],
        mode='markers',
        marker=dict(
            size=2,
            color='blue',
        )
    ), go.Scatter3d(
        x = x[200:,0],
        y = x[200:,1],
        z = y[0:200],
        mode='markers',
        marker=dict(
            size=2,
            color='red',
        )
    )]
)

# Graph data
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```

```python
x_3d = np.zeros([400,2])

for i in range(20):
    x_3d[i*20:(i+1)*20,0] = i/20
    x_3d[i*20:(i+1)*20,1] = np.linspace(0,100,20)/100.0

z = np.power(x_3d[:,0],3) + np.power(x_3d[:,1],3)

fig = go.Figure(
    data=[
        go.Mesh3d(
            x=x_3d[:,0],
            y=x_3d[:,1],
            z=z,
            opacity=0.5,
            color='blue'
        )
        ]
)

# Graph data
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```
