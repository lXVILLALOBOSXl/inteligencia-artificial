## 1. Introducción a Numpy
Operaciones con arreglos
Realizar en Numpy las operaciones de matrices siguientes:

A = 
\begin{bmatrix} -1 & 0 \\ 2 & 4 \end{bmatrix}^T \begin{bmatrix} 2 \\ 1 \end{bmatrix} - \begin{bmatrix} 1 \\ 2 \end{bmatrix}

B =
\begin{bmatrix} 1 & 5 & 3 \\ -2 & 0 & 1 \\ 3 & -4 & 10 \end{bmatrix} \begin{bmatrix} -2 \\ 4 \\ 9 \end{bmatrix} + \begin{bmatrix} 12 \\ 6 \\ -3 \end{bmatrix}

```python
import numpy as np

# Define the matrices
A = np.array([[-1, 0], [2, 4]])  # Original matrix
B = np.array([[2], [1]])  # Second matrix for multiplication
C = np.array([[1], [2]])  # Subtraction matrix

# Transpose A
A_T = A.T

# Perform the matrix multiplication and subtraction
A = A_T*B - C

# Print the result
print("A:\n", A) 

A2 = np.array([[1, 5, 3], [-2, 0, 1], [3, -4, 10]]) 
B = np.array([[-2], [4], [9]])  
C = np.array([[12], [6], [-3]])  

B = A2*B + C

print("B:\n", B) 
```

Realizar los reemplazos indicados en los arreglos siguientes:

C =
\begin{bmatrix} 1 & 0 & 2 \\ 2 & -4 & 3 \\ 9 & 7 & -1 \end{bmatrix} A(1,:) = \begin{bmatrix} 5, & -5, & 0 \end{bmatrix}

D =
\begin{bmatrix} 3 & 0 & 1 \\ 0 & 2 & 3 \\ -3 & 7 & 9 \end{bmatrix} A(2,:) = \begin{bmatrix} 4, & -2, & 3 \end{bmatrix}

```python
C = np.array([[1, 0, 2], [2, -4, 3], [9, 7, -1]]) 
C[1,:] = np.array([5, -5, 0])
print("C:\n", C)

D = np.array([[3, 0, 1], [0, 2, 3], [-3, 7, 9]]) 
D[2,:] = np.array([4, -2, 3])
print("D:\n", D)
```

A partir de los arreglos creados anteriormente, realizar las siguientes operaciones:

$A(:, 0)^T * B(0:2, 0:2) + D(0, 0:2)$

```python
result = np.dot(A[:,0].T, B[0:2,0:2]) + D[0,0:2]
print("Result:\n", result)
```

$B(1 : 2, 1 : 2) * B(0 : 1, 0 : 1)$

```python
result = np.dot(B[1:2,1:2], B[0:1,0:1])
print("Result:\n", result)
```

$D(:, 0) + C(:, 2) − B(0, :)^T$

```python
result = D[:,0] + C[:,2] - B[0,:].T
print("Result:\n", result)
```

$C(0 : 2, :)^T * A + B(1 : 3, 1 : 3)$

```python
result = np.dot(C[0:2,:].T,A) + B[1:2,1:3]
print("Result:\n", result)
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

y1 = np.sin(x)
y2 = -2.0*np.cos(x)
y3 = np.tan(x)

plt.figure(0, figsize=(7, 5))
plt.plot(x, y1, 'r', label='sin(x)')
plt.plot(x, y2, 'b', label='-2cos(x)')
plt.plot(x, y3, 'g', label='tan(x)')
plt.legend(loc='best')
plt.grid()
plt.title('Trigonometric functions')
plt.xlabel('Time(s)')
plt.ylabel('Signal')

plt.ylim(-10, 10)
plt.yticks(np.arange(-10, 10.1, 2.5))

plt.show()
```

### 2.1. Gráficas tipo Scatter
    Generar los clusters de puntos indicados a continuación (c = centro, σ = desviación estándar):
        A - Cluster de 200 puntos (c = (0,5, 0,8), σ = 0.2 )
        B - Cluster de 200 puntos (c = (0,3, 0,7), σ = 0.2 )
        C - Cluster de 200 puntos (c = (0,4, 0,4), σ = 0.15 )
        D - Cluster de 200 puntos (c = (0,6, 0,6), σ = 0.15 )

```python
# Create arrays
x = np.zeros([800, 2])  # 800 points, 2D space (for 4 clusters of 200 points each)
y = np.zeros(800)       # Labels for points

# Cluster A: Center (0.5, 0.8), σ = 0.2
x[:200, 0] = 0.5 + 0.2 * np.random.rand(200) - 0.2 * np.random.rand(200) # Cluster A along x-axis
x[:200, 1] = 0.8 + 0.2 * np.random.rand(200) - 0.2 * np.random.rand(200) # Cluster A along y-axis
y[:200] = np.zeros(200)  # Label for cluster A (class 0)

# Cluster B: Center (0.3, 0.7), σ = 0.2
x[200:400, 0] = 0.3 + 0.2 * np.random.rand(200) - 0.2 * np.random.rand(200)# Cluster B along x-axis
x[200:400, 1] = 0.7 + 0.2 * np.random.rand(200) - 0.2 * np.random.rand(200) # Cluster B along y-axis
y[200:400] = np.ones(200)  # Label for cluster B (class 1)

# Cluster C: Center (0.4, 0.4), σ = 0.15
x[400:600, 0] = 0.4 + 0.15 * np.random.rand(200) - 0.15 * np.random.rand(200) # Cluster C along x-axis
x[400:600, 1] = 0.4 + 0.15 * np.random.rand(200) - 0.15 * np.random.rand(200) # Cluster C along y-axis
y[400:600] = np.zeros(200)  # Label for cluster C (class 2)

# Cluster D: Center (0.6, 0.6), σ = 0.15
x[600:, 0] = 0.6 + 0.15 * np.random.rand(200) - 0.15 * np.random.rand(200)# Cluster D along x-axis
x[600:, 1] = 0.6 + 0.15 * np.random.rand(200) - 0.15 * np.random.rand(200) # Cluster D along y-axis
y[600:] = np.ones(200) # Label for cluster D (class 3)
```

Posteriormente, obtener las gráficas mostradas en las figuras siguientes:

**a) Cluster A vs B**

```python
# Plot data: Cluster A (class 0) vs Cluster B (class 1)
plt.figure(1, figsize=(5, 5))
plt.scatter(x[:400, 0], x[:400, 1], c=y[:400], label='Cluster A', marker='o', s=20)  # Cluster A in red

plt.grid()
plt.title('A Cluster vs B Cluster')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.show()
```

**a) Cluster C vs D**

```python
# Plot data: Cluster B (class 2) vs Cluster D (class 3)
plt.figure(2, figsize=(5, 5))
plt.scatter(x[400:,0], x[400:,1], c=y[400:], marker='o', s=20, cmap=plt.cm.Spectral)

plt.grid()
plt.title('C Cluster vs D Cluster')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.show()
```

### 2.2. Gráficas en 3D
    Generar los arreglos de datos siguientes:
        Generar una gradilla de puntos bidimensional X = (x, y), con valores entre 0 y 1.
        Calcular la función: 
$z = (x − 0,5)^3 + (y − 0,5)^2$

```python
# Generate the grid of points
x = np.linspace(0, 1, 20)
y = np.linspace(0, 1, 20)
X, Y = np.meshgrid(x, y)

# Calculate Z based on the function provided
Z = (X - 0.5)**3 + (Y - 0.5)**2
```

Posteriormente, generar una gráfica en 3 dimensiones.

```python
# Create a 3D plot
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')

# Plot the scatter plot
scatter = ax.scatter(X, Y, Z, c=X, s=10)

# Adjust the view angle (elevation, azimuth)
ax.view_init(elev=10, azim=215) 

# Show plot
plt.show()
```

## 3. Programación Orientada a Objetos
### 3.1. Herencia sen Python
Usando el paradigma de la programación orientada a objetos en python, crear las clases
indicadas a continuación. Posteriormente, crear una instancia de cada una y modificar sus
atributos usando los métodos disponibles.

![Figura 5: Herencia en Python.](./Python_Heritage.png)

a) Personal

**Atributos**
- **Nombre**: Cadena de texto con el nombre del empleado.
- **Edad**: Entero indicando la edad del empleado.
- **CURP**: Cadena de texto con el CURP del empleado.

**Métodos**
- **ModNombre(texto)**: El método permite modificar el nombre de la instancia.
- **ModEdad(número)**: El método permite modificar la edad de la instancia.
- **ModCURP(texto)**: El método permite modificar el CURP de la instancia.

---

b) Alumno (clase hija de Personal)

**Atributos**
- **Carrera**: Cadena de texto indicando la carrera del alumno.
- **Semestre**: Entero indicando el semestre actual del alumno.

**Métodos**
- **ModCarrera(texto)**: El método permite modificar la carrera del alumno.
- **ModSemestre(número)**: El método permite modificar el semestre del alumno.

---

c) Profesor (clase hija de Personal)

**Atributos**
- **Departamento**: Cadena de texto indicando el departamento donde trabaja el profesor.

**Métodos**
- **ModDepartamento(texto)**: El método permite modificar el departamento del profesor.

---

Posteriormente, crear instancias para cada una de las clases y hacer un llamado a cada uno de sus métodos.


```python
class Staff:
    def __init__(self, name, age, CURP):
        self.name = name
        self.age = age
        self.CURP = CURP
        
    def mod_name(self, name):
        self.name = name

    def mod_age(self, age):
        self.age = age

    def mod_CURP(self, CURP): 
        self.CURP = CURP

class Student(Staff):
    def __init__(self, name, age, CURP, career, semester):
        super().__init__(name, age, CURP)
        self.career = career
        self.semester = semester

    def mod_career(self, career):
        self.career = career
    
    def mod_semester(self, semester):
        self.semester = semester

class Teacher(Staff):
    def __init__(self, name, age, CURP, department):
        super().__init__(name, age, CURP)
        self.department = department    

    def mod_department(self, department):
        self.department = department


def main():

    person = Staff("Juan", 25, "JUJU123456HDFR")
    print(f'\nBefore\nName: {person.name}\nAge: {person.age}\nCURP: {person.CURP}')
    person.mod_name("Pedro")
    person.mod_age(30)
    person.mod_CURP("PEPE123456HDFR")
    print(f'After\nName: {person.name}\nAge: {person.age}\nCURP: {person.CURP}')
    
    student = Student("Maria", 20, "MAMA123456HDFR", "Computer Science", 3)
    print(f'\nBefore\nName: {student.name}\nAge: {student.age}\nCURP: {student.CURP}\nCareer: {student.career}\nSemester: {student.semester}')
    student.mod_career("Software Engineering")
    student.mod_semester(5)
    print(f'After\nName: {student.name}\nAge: {student.age}\nCURP: {student.CURP}\nCareer: {student.career}\nSemester: {student.semester}')

    teacher = Teacher("Luis", 35, "LULU123456HDFR", "Engineering")
    print(f'\nBefore\nName: {teacher.name}\nAge: {teacher.age}\nCURP: {teacher.CURP}\nDepartment: {teacher.department}')
    teacher.mod_department("Mathematics")
    print(f'After\nName: {teacher.name}\nAge: {teacher.age}\nCURP: {teacher.CURP}\nDepartment: {teacher.department}')

if __name__ == "__main__":
    main()
    
```
